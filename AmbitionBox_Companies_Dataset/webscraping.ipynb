```python
import pandas as pd
import requests
import re
from bs4 import BeautifulSoup
```

```python
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'} 
webpage = requests.get('https://www.ambitionbox.com/list-of-companies?page=1',headers=headers).text
```

```python
soup = BeautifulSoup(webpage,'lxml')
```

```python
company=soup.find_all('div',class_="companyCardWrapper__primaryInformation")
```

```python
len(company)
```

```python
name=[]
rating=[]
reviews=[]
type_and_location=[]


for i in company:
    name.append(i.find('h2').text.strip())
    rating.append(i.find('div',class_='rating_text').text.strip())
    reviews.append(i.find('span', class_="companyCardWrapper__companyRatingCount").text.strip())
    type_and_location.append(i.find('span',class_='companyCardWrapper__interLinking').text.strip())
d = {'name':name,'rating':rating,'reviews':reviews,'type_and_location':type_and_location}
df = pd.DataFrame(d)
```

```python
df
```

```python
final = pd.DataFrame()

for j in range(1,100):
    url='https://www.ambitionbox.com/list-of-companies?page={}'.format(j)

    headers = {'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'} 
    webpage = requests.get(url,headers=headers).text
    
    soup = BeautifulSoup(webpage,'lxml')
    
    company=soup.find_all('div',class_="companyCardWrapper__primaryInformation")
    name=[]
    rating=[]
    reviews=[]
    type_and_location=[]
    
    
    for i in company:
        
        name.append(i.find('h2').text.strip())
        
        rating.append(i.find('div',class_='rating_text').text.strip())
        reviews.append(i.find('span', class_="companyCardWrapper__companyRatingCount").text.strip())
        type_and_location.append(i.find('span',class_='companyCardWrapper__interLinking').text.strip())
    d = {'name':name,'rating':rating,'reviews':reviews,'type_and_location':type_and_location}
    df = pd.DataFrame(d)
    final = pd.concat([final, df], ignore_index=True)


    
    
```

```python
final.to_csv("ambitionbox_companies_dataset.csv", index=False)
```
